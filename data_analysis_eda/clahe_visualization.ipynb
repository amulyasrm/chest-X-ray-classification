# 🏥 CLAHE Preprocessing Visualization
This notebook demonstrates the impact of **Contrast Limited Adaptive Histogram Equalization (CLAHE)** on chest X-ray images. CLAHE is used to enhance local contrast, making subtle pathological features (like nodules or lung textures) more visible for the deep learning model.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Path to a sample image
image_path = '/Users/amulya/Documents/capstone/data/images/00000001_000.png'

def apply_clahe(img_path):
    # Load image in grayscale
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    
    # Initialize CLAHE
    # clipLimit is the threshold for contrast limiting
    # tileGridSize defines the number of tiles in row and column
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    img_clahe = clahe.apply(img)
    
    return img, img_clahe

# Process
original, enhanced = apply_clahe(image_path)

# Visualization
plt.figure(figsize=(15, 7), facecolor='#FAFAFA')

# Original Subplot
plt.subplot(1, 2, 1)
plt.imshow(original, cmap='gray')
plt.title('Low Contrast (Original)', fontsize=14, pad=15, fontweight='bold', color='#333333')
plt.axis('off')

# CLAHE Subplot
plt.subplot(1, 2, 2)
plt.imshow(enhanced, cmap='gray')
plt.title('High Contrast (Processed with CLAHE)', fontsize=14, pad=15, fontweight='bold', color='#D4AF37')
plt.axis('off')

plt.tight_layout()
plt.show()
```
